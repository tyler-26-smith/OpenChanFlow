## Problem 4

Develop a Manning equation solver that will return the normal depth of a trapezoidal channel given the discharge ($Q$), channel bottom width ($b$), slope ($S$), side slope ($m$ where $m$ is the run per unit rise), and the Manning $n$ value. Use the function ``fsolve`` to do this. You will need to import this function from the scipy library by using the following statement: ``from scipy.optimize import fsolve``. You can import this anywhere above the place where you call ``fsolve``. I like to put all of my imports just below the ``%pylab inline``statement near the top of the file. Test your solver by printing the normal depth for the following conditions: $Q=$ 350 cms, $b=$ 100 m, $S=0.0003$, $n=0.04$, with $m=2$. You should get that the depth, $y$ is equal to 3.47 m.

```python
%pylab inline 
from scipy.optimize import fsolve 

def mannings(f, Q, b, S, n, m):
    return(Q-((1/n)*(b*f+m*f*f)*(((b*f+m*f*f)/(b+2*f*sqrt(1+m**2)))**(2/3))*sqrt(S)))

# Input conditions
Q = 350
b = 100
S = 0.0003
n = 0.04
m = 2

# f guess 
fguess = 2

# fsolve 
f_solution = fsolve(mannings,fguess, args=(Q,b,S,n,m))

# solution
print('The normal depth y is =', f_solution)
print('The normal depth y is =', around(f_solution,decimals=2))
```
