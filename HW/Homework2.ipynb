## Question 1
**Problem 1.1 of Moglen (2015)**

What slope would lead to a 1% difference between depth in the vertical plane rather than depth measured perpendicular to the channel bottom? Compare this slope to the observation that a channel slope
of $S_0 = 0.01 m/m$ is generally considered quite steep for open channel
flow.

```python
import numpy as np
angle_rad = np.arccos(1/(1.01))
angle_deg = np.rad2deg(angle_rad)
slope = np.tan(angle_rad)

print('angle (degrees) =', round(angle_deg, 2))
print('slope (m/m) =', round(slope, 2))
```

A difference of 1% in the vertical plane and the depth perpendicular to the channel bed would lead to a channel slope of 0.14 (m/m). This small difference would create a slope 14 times greater than that of a channel slope of 0.01 m/m.

## Question 3
Follow the derivation for the celerity of a shallow, small amplitude wave in
chapter 1 of Moglen (2015). Explicitly show all of the steps that lead up to
equation 1.13. Are there any assumptions that seem questionable?

Assuming rectangular channel with width (w) and applying continuity equation

$$
Q_1 = Q_2
$$

$$
v_1*y_1*w = v_2*y_2*w
$$

Dividing both sides of equation by w and substituting $y_{1,2}$ for $v_{1,2} = c - \Delta v$

$$
(c-\Delta v)*(y + \Delta y) = c *y
$$

$$
c*y+c*\Delta y - \Delta v * y - \Delta v * \Delta y = c*y
$$

$\Delta v$ is much smaller than c, and $\Delta y$ is smaller to y and therefore $\Delta v * \Delta y$ can be neglected

$$
\frac{c}{y} = \frac{\Delta v} {\Delta y}
$$

$\Delta y$ is the wave height, and since it is "small", it is treated as negligable. Using Bernoulli's equation, and assuming the pressure remains constant at the begining and ending cross sections

$$
y + \Delta y + \frac{(c-\Delta v)^2}{2g}= y + \frac{c^2}{2g}
$$

$$
y + \Delta y + \frac{c^2-2c*\Delta v +\Delta v^2}{2g}= y + \frac{c^2}{2g}
$$

Again since $\Delta v$ is small $\Delta v^2$ can be neglected on both sides of the equation

$$
\Delta y - \frac{c*\Delta v}{g} = 0
$$

$$
\frac{g}{c} = \frac{\Delta v}{\Delta y}
$$

Since $\frac{\Delta v} {\Delta y} = \frac{c}{y} = \frac{g}{c}$

$$
\frac{c}{y}=\frac{g}{c}
$$

$$
\boxed{ 
c = \sqrt{gy}
}
$$

Looking at the work shown in Moglen, I beleive making the assumption that $\Delta v$ and $\Delta y$ are negligable when they are either multiplied or squared could potentially be problematic. If the shallow wave formula is also applied to an area that is not very deep to begin with, the waves depth and celerity become proportionally larger relative to the channels depth and velocity. If this were to happen the prior assumptions are no longer valid in a system of this nature.

## Question 5
Consider a river with an average bed slope of, $S_o = 0.0002$, a depth of $h = 5 m$ and width of $100 m$. Assume that the flow is **uniform**, and that the velocity distribution over the vertical (the z direction) can be adequately described from $\frac{z}{h} = 0.001$ to 1 by: 

$$
\frac{u}{u_o} = \frac{1}{k} ln(\frac{z}{z'})+8.5
$$

where $u_o$ is the friction or shear velocity, $ = 0.4$ is the von Karman constant, and $z'$ is the roughness length scale. For this problem, take $z' = 0.05$, and assume that the uniform flow equation $u_0 = \sqrt{ghS_o}$ is valid. Use this information to calculate the following:

(a) volumetric flow rate, Q

(b) depth-average velocity, V

(c) Reynolds number, Re

(d) Froude number, Fr

State any assumptions you make in the process

#### (a) volumetric flow rate, Q
$$
Q = qw 
$$

$$
q= \int_{z}^{} u \,dz\
$$

Limits of integration: z/h = 0.001 to 1 or z= 0.005 to 5

$$
q= \int_{0.005}^{5} u \,dz\
$$

$$
u = \frac{\sqrt{ghS_o}}{k} ln(\frac{z}{z'})+8.5\sqrt{ghS_o}
$$

$$
q = \int_{0.005}^{5} \frac{\sqrt{ghS_o}}{k} ln(\frac{z}{z'})+8.5\sqrt{ghS_o} \, dz\
$$

$$
Q = qw
$$

**Assuming rectangular channel area** $A=wh= 5m(100m)= 500m^2$

```python
from scipy.integrate import quad
import numpy as np

# Define Parameters 
g = 9.81
h = 5
S = 0.0002
k = 0.4
z = 0.05
w= 100

# definning function
def integrand(x):
    return (np.sqrt(g * h * S) / k) * np.log(x / z) + 8.5 * np.sqrt(g * h * S)

# integrating equation
q_5, error = quad(integrand, 0.001, 1)

# Print q
#print("Result of integration:", q_5)
#print("Estimated absolute error:", error)

# Solving for Q
Q5 = q_5*w
print("Q (m/s^2)=", Q5)
```

#### (b) depth-average velocity, V
$$
u = \frac{\sqrt{ghS_o}}{k} ln(\frac{z}{z'})+8.5\sqrt{ghS_o}
$$

```python
print("depth averaged velocity (m/s)=", round(q_5, 2))
```

#### (c) Reynolds number, Re
$$
R_e = \frac{vg}{v}
$$

```python
u = q_5
visc = 1.004*10**(-6) #m2/s
g = 9.81

R_e = (u*g)/visc

print("Renyolds Number =", round(R_e, 2))
```

#### (d) Froud number, Fr
$$
F_r = \frac{v}{\sqrt{gy}}
$$

```python
u = q_5
g = 9.81
y = 5

F_r = (u/((g*y)**0.5))

print("Renyolds Number =", round(F_r, 2))
```
