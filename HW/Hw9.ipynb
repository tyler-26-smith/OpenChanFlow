What is the difference between hydrologic and hydraulic routing?

<span style="font-size: larger;">**<u>Hydrologic Routing</u> -**</span> Derived from the continuity equation applies the volume of water between x_{in} & x_{out} by integrating over x, *does not consider force.*

$\Downarrow$ <u>**Continuity Equation:**<u/>

$$
\boxed{
\frac{dA}{dt} + \frac{dQ}{dx} = 0
}
$$

$\downarrow$

$$
\int_{x_{in}}^{x_{out}} \frac{dA}{dt} \, dx + \int_{x_{in}}^{x_{out}} \frac{dQ}{dx} \, dx = 0
$$

$\Downarrow$ <u>**Governing Equation:**</u>

$$
\boxed{
\frac{dS}{dt} = I - O
}
$$

$\downarrow$ Discritize


$$
\frac{S^{k+1} - S^k}{\Delta t} \approx \frac{dS}{dt}
$$

$\downarrow$
    
$I = \frac{1}{2} \cdot (I^k + I^{k+1})$
    
$O = \frac{1}{2} \cdot (O^k + O^{k+1}$
    
**k = indicies**
    
$S^k =$ **storage at current time step**
    
$S^{k+1} =$ **storage at next timestep**
    
$\Delta t =$ **time change**

$$
\frac{S^{k+1} - S^{k}}{\Delta t} = \frac{1}{2} \cdot \left( I^k + I^{k+1} \right) - \frac{1}{2} \cdot \left( O^k + O^{k+1} \right)
$$

$\Downarrow$ <u>**Solution Equation 1:**</u>

$$
\boxed{
\frac{2S^{k+1}}{\Delta t} + O^{k+1} = I^k + I^{k+1} - O^k + \frac{2S^k}{\Delta t}
}
$$

$\downarrow$ relating unknowns
    
$S = S_w + S_p$
    
    - S_w = storage in wedge
    
    - S_p = storage in prism
$S_p = S_p(O)$
    
$S_w = S_w(I, O)$

$S = \theta(O)$
    
$X = \frac{I}{O}$
    
$\Downarrow$ <u>**Solution Equation 2 (Muskingum):**</u>
    
$$
\boxed{
S = \theta \left[X \cdot I + \left(1 + X \right) \cdot O \right]
}
$$

<span style="font-size: larger;">**<u>Hydraulic Routing</u> -**</span> Derived from the continuity *and* momentum equation, *does consider force.* Solved by plugging momentum equation into mannings equation then finally converting to continuity. There are three types of Hydrologic routing, the type of routing is dependant on the amount of the momentum equation used:

    1. Kinematic Wave Routing
    2. Diffusion Wave Routing
    3. Dynamic Wave Routing

$\Downarrow$ <u>Full Momentum Equation:</u>

$$
\boxed{
\frac{dv}{dt} + v \frac{dv}{dx} + g \frac{dy}{dx} - g \left(S_o - S_f \right) = 0
}
$$

<u>Kinematic Wave:</u>   $-g(S_o - S_f) = 0$

<u>Diffusion Wave:</u>   $ g \frac{dy}{dx} -g(S_o - S_f) = 0$

<u>Dynamic Wave:</u>   $\frac{dv}{dt} + v \frac{dv}{dx} g \frac{dy}{dx} -g(S_o - S_f) = 0$

$\downarrow$ Reduced momentum equation (kinematic):

$$
S_f = S_o
$$

$\downarrow$ continuity equation:

$$
\frac{dA}{dt} + \frac{dQ}{dx} = 0
$$

$\downarrow$ Put into manning's equation:

$$
Q = \frac{c_o}{k} A R^{2/3}{S_f}^{1/2}
$$

$\Downarrow$ <u>**Governing Equation (kinematic here, can be applied to routing types):**</u>

$$
\boxed{
\frac{dQ}{dt} + c_k + \frac{dQ}{dx} = 0
}
$$

c_k = kinematic wave velocity

$\downarrow$ discritize:

$\frac{dQ}{dt} + c_k + \frac{dQ}{dx} = 0$  $\rightarrow$ $A + B =0$

**<u>A:</u>**

$$
\frac{dQ}{dt} = \frac{X \left( {Q_i}^{k+1} - {Q_i}^k \right) \cdot (1 - X) \cdot \left( {Q_{i+1}}^{k+1} - {Q_i}^k \right)}{\Delta t}
$$

**<u>B:</u>**

$$
c_k \frac{dQ}{dx} = c_k \left[ \frac{Y \left( {Q_{i+1}}^{k} - {Q_i}^{k} \right) \cdot (1 + Y) \cdot \left( {Q_{i+1}}^{k} - {Q_i}^k \right)}{\Delta x} \right]
$$

$\Downarrow$ <u>**In Channel Routing Form:**</u>

$$
\boxed{
{Q_{i+1}}^{k+1} = C_o {Q_i}^{k+1} + C_1 {Q_i}^k + C_2 {Q_{i+1}}^k
}
$$


$C_o = \frac{-X + C_n \cdot (1 - Y)}{1 - X + C_n \cdot (1 - Y)}$

$C_1 = \frac{X + C_n \cdot Y}{1 - X + C_n \cdot (1 - Y)}$

$C_2 = \frac{(1 -X) - C_n \cdot Y}{1 - X + C_n \cdot (1 - Y)}$

$C_n = c_k \frac{\Delta t}{\Delta x}
    
    C_n = courant number

$\downarrow$ If Y = 0.5 $\rightarrow$$\theta = \frac{\Delta x}{c_k}$$\rightarrow$ use muskingum

$Y = 0.5$

$X = 0.5 \left(1 - \frac{Q}{B \cdot S_o \cdot c_k \cdot \Delta x} \right)$

    B = topwidth

### Equations

```python
%matplotlib inline
from scipy.integrate import odeint, ode
from scipy.optimize import fsolve, newton
import pandas as pd

def manning(y, Q, b, S, n, g, m):
    if g>10:
        Co = 1.486
    else: 
        Co= 1
    A = (b + m * y) * y
    R = A / (b + 2 * y * (1 + m**2)**(1/2))
    return ((Co/n)*A*R**(2/3)*S**(1/2)) - Q

def area(y):
    return (b+m*y)*y

def hydrRad(y):
    return area(y) / (b+2*y*(1+m**2)**(1/2))

def topwidth(y):
    return b+2*y*m
```

### Muskingum-Cunge

```python
Qbase = 500
Qpeak = 4500

tpeak_hr = 2
tbase_hr = 4

Tsim_hr = 24 

So = 0.0005
b = 100
m = 2
n = 0.03
L_mi = 30

g = 32.2

print(Qbase)
```

### Estimate Celerity, Ck

```python
Q = (Qbase + Qpeak)/2
#find normal depth
yo = fsolve(manning, 1, args=(Q, b, So, n, g, m))
Vo = Q/area(yo)
Ck = 5*Vo/3
B = topwidth(yo)
print("Conditions at mid Discharge:")
print("- normal depth", around(yo, decimals=3))
print("- topwidth", around(B, decimals=3))
print("- celerity", around(Ck, decimals = 3))
```

### Chose a delta t

```python
dt_hr = 0.5
```

### Calculate the max delta x

```python
dt = dt_hr * 60 * 60
dx_max = 0.5*(Ck * dt + Q / (B*So*Ck))
print("max space step dx", around(dx_max, decimals=3))
```

### Calculate delta x and the number of reach segments

```python
L = L_mi * 5280
Nri = L/dx_max
Nr = int(ceil(Nri))

dx = L/Nr

Cn = Ck * dt/dx
X = 0.5*(1-Q/(B*So*Ck*dx))
print("number of reaches =", Nr, ", dx =", dx)
print("Courant Number =", around(Cn, decimals=2))
print("X", around(X, decimals=2))
```

### Calculate the coefficents for routing

```python
C0 = (0.5 * Cn - X)/(1-X+0.5*Cn)
C1 = (0.5 * Cn + X)/(1-X+0.5*Cn)
C2 = (1 - X - 0.5*Cn)/(1-X+0.5*Cn)
print("C0 =", around(C0, decimals=3), "C1 =", around(C1, decimals=3), "C2 =", around(C2, decimals =3))
print("Unity Check: C0 + C1 + C2 =", C0+C1+C2, "Should be equal to 1")
```

### Develope inflow hydrograph

```python
t_hr = arange(0., Tsim_hr+dt_hr, dt_hr)
Qinflow = zeros(len(t_hr))

Qinflow[0] = Qbase

for i in range(1, len(Qinflow)):
    if t_hr[i] <= tpeak_hr:
        Qinflow[i] = Qinflow[i-1] + ((Qpeak-Qbase)/(tpeak_hr))*dt_hr
    elif t_hr[i] <= tpeak_hr+tbase_hr:
        Qinflow[i] = Qinflow[i-1]-((Qpeak-Qbase)/(tbase_hr))*dt_hr
    else:
        Qinflow[i]=Qbase
        
plot(t_hr, Qinflow, "o-")
xlabel("Time (hr)")
ylabel("Discharge, $Q$ (cfs)");
```

### Run the Muskingum-Cunge routing. Output from one reach is the input for the next

```python
Qin = Qinflow
Qout = zeros(len(Qin))
Qout[0] = Qbase

for j in range(0,Nr):
    for i in range(0, len(Qin)-1):
        Qout[i+1]=C0*Qin[i+1]+C1*Qin[i]+C2*Qout[i]
    if j < Nr -1:
        Qin = copy(Qout)
        
plot(t_hr, Qinflow, "b", label="Inflow")
plot(t_hr, Qout, label="Outflow")
xlabel("Time (hr)")
ylabel("Discharge, $Q$ (cfs)")
legend(loc=1);
```

### Calculate key values such as peak discharge at the outlet and time for the wave to move through the reach

```python
i_peak = argmax(Qout)
travel_hr = t_hr[i_peak] - tpeak_hr
Qreduc = Qpeak-Qout[i_peak]

print("Key Values:")
print("- Propigation Time To Downstream =", around(travel_hr, decimals=2))
print("- Reduction in Peak Discharge =", around(Qreduc, decimals=2))
```
